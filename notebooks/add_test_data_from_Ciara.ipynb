# Add hand-transcribed test data from Ciara

Ciara Ryan <ciara.ryan@met.ie> has already transcribed and QCd some of the Irish daily rainfall sheets during her PhD. These have the same format as the UK ones I'm looking at.

She's sent me her results for 64 images. These are known-good transcriptions, not really enough to be a training dataset, but ideal for validation.




## 1. Getting the data in the format I'm using

I've downloaded the data to ```test_data/from_Ciara/originals```.

It needs reformatting to match the images/transcriptions files I'm using here. The process needs software for pdf and excel file manipulation, so there's a special software environment for that.

```bash
conda env create -f test_data/from_Ciara/ciara_convert.yml
conda activate ciara-convert
python scripts/convert_ciara_test_data.py
```

This makes ```test_data_from_Ciara/images``` and ```test_data_from_Ciara/transcriptions``` to match all the other examples.


## 2. Upload the new test data to Azure

```bash
conda activate weather-doc-extractor
az login
bash scripts/aml_upload.sh --src test_data/from_Ciara --dst test_data/from_Ciara
```


## 3. Run a test extraction on the new images

```bash
bash scripts/aml_submit.sh extract \
  --images-path test_data/from_Ciara/images \
  --model granite4 \
  --limit 20 \
  --batch-size 8
```
